1. Select only the transactions that have occurred during the period 09/01/11 and 11/30/11 and create a new subset of data. For answering Questions 2 to 5, use this subset of data.

```python
input_df['InvoiceDate'] = pd.to_datetime(input_df['InvoiceDate'], dayfirst=True)

start_date = pd.to_datetime('2011-09-01')
end_date = pd.to_datetime('2011-11-30')

filtered_df = input_df[(input_df.InvoiceDate >= start_date) & (input_df.InvoiceDate <= end_date)]
filtered_df
```

2. Transform the above dataset in Q1 into another dataset where each record represents an invoice. The record will have an `InvoiceNo` as a column and `StockCodes` of the corresponding items bought in the invoice. THe number of `StockCodes` in an invoice can be variable.

```python
filtered_df = filtered_df.groupby('InvoiceNo')['StockCode'].apply(list).reset_index().rename(columns={'StockCode': 'StockCodes'})
filtered_df
```

3. Prepare the data and generate the association rules from the above dataset in Q2. Filter out all the rules by minimum support of 0.01 and lift of more than 1.0.

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

transactions = filtered_df['StockCodes'][0:5]

t_encoder = TransactionEncoder()
t_encoder_array = t_encoder.fit(transactions).transform(transactions)
encoded_df = pd.DataFrame(t_encoder_array, columns=t_encoder.columns_)

frequent_itemsets = apriori(encoded_df, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
filtered_rules = rules[(rules['support'] >= 0.01) & (rules['lift'] > 1.0)]
filtered_rules
```

4. Find the top 10 rules from the above association rule set, sorted by confidence in descending order.

```python
top_10_rules = filtered_rules.sort_values(by='confidence', ascending=False).head(10)
top_10_rules
```

```python
ratings_df = pd.read_csv('../clustering-datasets/ratings.csv')
ratings_df.head(5)
```

6. Filter out the ratings records for the movies that belong to only “Action” genre. Calculate item simi-larity indexes between the movies based on movies the users have bought in “Action” genre and how they have rated them. Use Pearson correlation coefficient to find similarities. Then recommend top 5 similar movies to the following movies:
    - Heat
    - Eraser

```python
action_movies = movies_df[movies_df['genres'].str.contains('Action')]
action_ratings = ratings_df[ratings_df['movieId'].isin(action_movies['movieId'])]

from scipy.stats import pearsonr

def calculate_similarity(movie_id_1, movie_id_2):
    ratings1 = action_ratings[action_ratings['movieId'] == movie_id_1]
    ratings2 = action_ratings[action_ratings['movieId'] == movie_id_2]
    
    common_users = set(ratings1['userId']).intersection(set(ratings2['userId']))
    if len(common_users) == 0:
        return 0  # No common users, so correlation is 0
    
    ratings1 = ratings1[ratings1['userId'].isin(common_users)]
    ratings2 = ratings2[ratings2['userId'].isin(common_users)]
    
    if len(ratings1) == 0 or len(ratings2) == 0:
        return 0  # Not enough data to calculate correlation
    
    corr, _ = pearsonr(ratings1['rating'], ratings2['rating'])
    return corr

movie_to_recommend = ['Heat', 'Eraser']

for movie in movie_to_recommend:
    movie_id_matches = movies_df[movies_df['title'] == movie]
    
    if not movie_id_matches.empty:
        movie_id = movie_id_matches['movieId'].values[0]
        similar_movies = []

        for other_movie_id in action_movies['movieId']:
            if other_movie_id != movie_id:
                similarity = calculate_similarity(movie_id, other_movie_id)
                similar_movies.append((other_movie_id, similarity))
        
        similar_movies.sort(key=lambda x: x[1], reverse=True)
        top_similar_movies = similar_movies[:5]
        
        print(f"Recommendations for {movie}:")
        for top_movie_id, similarity in top_similar_movies:
            movie_title = movies_df[movies_df['movieId'] == top_movie_id]['title'].values[0]
            print(f"{movie_title} (Similarity: {similarity:.4f}")
        print("\n")
    else:
        print(f"{movie} not found in the dataset.\n")
```
